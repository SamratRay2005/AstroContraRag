```python
import matplotlib.pyplot as plt
import numpy as np

# ------------------ DATA ------------------
labels = ["Standard ($K=200$)", "Optimized ($K=20$)"]
recall_scores = [98.8, 96.6]        # Left Axis
candidate_counts = [258.7, 20.0]    # Right Axis
x = np.arange(len(labels))

# ------------------ CONFIG ------------------
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12
})

fig, ax1 = plt.subplots(figsize=(6.5, 4.5))

# ------------------ LEFT AXIS (Recall) ------------------
# Use 'hatch' (patterns) for grayscale safety
bars = ax1.bar(x, recall_scores, width=0.4, color='white',
               edgecolor='black', hatch='///', label='Recall (%)')
ax1.set_ylabel("Recall (%)")
ax1.set_ylim(90, 100)

# Label bars
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height - 1.5,
             f"{height:.1f}%", ha='center', va='top',
             color='black', fontweight='bold',
             bbox=dict(facecolor='white', edgecolor='none', pad=0.5))

# ------------------ RIGHT AXIS (Latency) ------------------
ax2 = ax1.twinx()
# Use distinct marker ('D' = Diamond) and dashed line
ax2.plot(x, candidate_counts, color='black', marker='D', markersize=8,
         linestyle='--', linewidth=1.5, label='Candidates Processed')
ax2.set_ylabel("Avg. Candidates Processed")
ax2.set_ylim(0, 300)

# Label points
for i, val in enumerate(candidate_counts):
    ax2.text(x[i], val + 15, f"{int(val)}", ha='center',
             color='black', fontweight='bold')

# ------------------ STYLING ------------------
ax1.set_xticks(x)
ax1.set_xticklabels(labels)
ax1.set_title("Efficiency vs. Effectiveness Trade-off")

# ------------------ SAVE ------------------
plt.tight_layout()
plt.savefig("fig4_efficiency.pdf", bbox_inches="tight")
plt.savefig("fig4_efficiency.png", dpi=600, bbox_inches="tight")
plt.show()
print("Saved Fig 4 as PDF and PNG.")
```
