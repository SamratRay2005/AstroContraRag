```python
import matplotlib.pyplot as plt
import numpy as np

# ------------------ DATA ------------------
total = 2336
valid = 2026
invalid = total - valid
accuracy = (valid / total) * 100

# ------------------ CONFIG ------------------
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "axes.spines.top": False,    # Remove top border
    "axes.spines.right": False   # Remove right border
})

fig, ax = plt.subplots(figsize=(5, 4)) # Compact vertical size

# ------------------ PLOT ------------------
categories = ['Valid Contradictions', 'Invalid / Hallucinations']
counts = [valid, invalid]
colors = ['#404040', '#D3D3D3'] # Dark Grey vs Light Grey

bars = ax.bar(categories, counts, color=colors, edgecolor='black', width=0.6)

# ------------------ ANNOTATIONS ------------------
# Place text directly above the bars
for bar, count in zip(bars, counts):
    height = bar.get_height()
    percent = (count / total) * 100

    # Label: "Count"
    ax.text(bar.get_x() + bar.get_width()/2, height + 50,
            f"{count}",
            ha='center', va='bottom', fontsize=12, fontweight='bold')

    # Label: "(Percentage)"
    ax.text(bar.get_x() + bar.get_width()/2, height - 150 if height > 200 else height + 150,
            f"({percent:.1f}%)",
            ha='center', va='top' if height > 200 else 'bottom',
            fontsize=11, color='white' if height > 200 else 'black')

# ------------------ STYLING ------------------
ax.set_ylabel("Number of Queries")
ax.set_title(f"LLM-as-a-Judge Validation ($n={total}$)", pad=20)
ax.set_ylim(0, total + 300) # Give space for text

plt.tight_layout()
plt.savefig("fig7_validation_clean.pdf", bbox_inches="tight")
plt.savefig("fig7_validation_clean.png", dpi=600, bbox_inches="tight")
plt.show()

print("Saved Cleaner Fig 7 (Vertical Bar)")
```
